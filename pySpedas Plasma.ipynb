### Importação do Plasmapy

* https://docs.plasmapy.org/en/stable/formulary/index.html

```python
#!pip install plasmapy
import plasmapy
```

### Velocidade de Alfvén 

$$
V_A = \dfrac{B}{\sqrt{μ_0 n_i}}
$$

```python
Va = plasmapy.formulary.speeds.Alfven_speed(B, n_i, 'p')

# convert to km / s
Va = Va.to(u.km / u.s)

Va
```

### Beta de Plasma

$$
β = \dfrac{p_{th}}{p_{mag}}
$$

```python
# ions
beta_i = plasmapy.formulary.dimensionless.beta(T_i, n_i, B)

# electrons
beta_e = plasmapy.formulary.dimensionless.beta(T_e, n_e, B)

# combined
beta = beta_i + beta_e

beta
```

### Ion inertial length

$$
d = \frac{c}{ω_{ps}}
$$

```python
d_i = plasmapy.formulary.lengths.inertial_length(n_i, 'p+')

d_i
```

## Comprimento de Debye 

$$
  \lambda_D = \sqrt{\dfrac{\epsilon_0 k_B T_e}{n_e q_e^2}}
$$

```python
lamda_d = plasmapy.formulary.lengths.Debye_length(T_e, n_e)
lamda_d
```

```python
lamda_di = plasmapy.formulary.lengths.Debye_length(T_i, n_i)
lamda_di
```

### Frequênca de giro dos Ions

$$
 ω_{c} = \frac{|Z| e B}{m}
$$

```python
omega_ci = plasmapy.formulary.frequencies.gyrofrequency(B, 'p', to_hz=True)
omega_ci
```

```python
omega_ce = plasmapy.formulary.frequencies.gyrofrequency(B, 'e', to_hz=True)
omega_ce
```

### Raio de larmor dos ions

$$
 r_{L,s} = \frac{γ m_s V_{⟂,s}}{ |q_s| B}
$$

```python
r_i = plasmapy.formulary.lengths.gyroradius(B, 'p', T=T_i)
r_e = plasmapy.formulary.lengths.gyroradius(B, 'e', T=T_e)

# convert to km
r_i = r_i.to(u.km)
r_e = r_e.to(u.km)

r_i
```

## Região Pré-shock 

```python
def IVal(v,init,fim):
    
    tam = len(v)
    t   = np.zeros((tam, 2))
    i   = np.zeros((1, 2))
    
    for ii in range(tam):
        
        t[ii, 0] = np.abs(init - v[ii])
        t[ii, 1] = np.abs(fim - v[ii])   
    
    i[0,0] = t[:,0].argmin()
    i[0,1] = t[:,1].argmin()
    
    return i
```

```python
tim = [t1_b, t1_di, t1_de, t1_vi, t_dis, t_jj, t_p1]

ips1 = np.zeros((len(tim),2))

for k in range(len(tim)):
    
    ips1[k] = IVal(tim[k], tb_i1, tb_s1)
```
